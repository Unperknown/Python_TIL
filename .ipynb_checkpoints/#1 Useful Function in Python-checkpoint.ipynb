# Useful Method in Python

Python 코드를 작성할 때 도움이 되는 메서드들을 모아놓았다.

```python
a = 0
type(a)
```

```python
help(a) #객체에 대한 정보 보여주기
```

```python
dir(a) # 객체의 필드와 메서드 목록
```

```python
a? #객체의 문서 보여주기
```

```python
id(a) #객체의 주소 보여주기
```

```python
import builtins
dir(builtins) #Python 내장 필드 및 메서드 보여주기
```
